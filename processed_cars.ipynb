```python
import pandas as pd
import numpy as np
```

```python
file_name=r'C:\Users\khush\Downloads\usedcars.csv'
```

```python
df=pd.read_csv(file_name,header=None)
```

```python
df.head(5)
```

```python
header_names=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns=header_names
df.head(5)
```

```python
df.dtypes
```

```python
df.describe(include="all")
```

```python
df=df.replace("?",np.nan)
df.head(10)
```

```python
df.info()
```

```python
#Pre-processing
```

```python
#missing vallues
#drop he values
df.dropna(subset=["price"],axis=0,inplace=True)
df.head(5)
```

```python
#evaluate
x=df.isnull()
x.head()
```

```python
#counting
for column in x.columns.values.tolist():
    print(x[column].value_counts())
    print("")
```

```python
mean_norm_loss=df["normalized-losses"].astype(float).mean(axis=0)
df["normalized-losses"]=df["normalized-losses"].fillna(mean_norm_loss)
df.head(5)
```

```python
mean_bore=df["bore"].astype(float).mean(axis=0)
df["bore"]=df["bore"].fillna(mean_bore)
df.head(5)
```

```python
mean_stroke=df["stroke"].astype(float).mean(axis=0)
df["stroke"]=df["stroke"].fillna(mean_stroke)
df.head(5)
```

```python
mean_horsepower=df["horsepower"].astype(float).mean(axis=0)
df["horsepower"]=df["horsepower"].fillna(mean_horsepower)
df.head(5)
```

```python
mean_peakrpm=df["peak-rpm"].astype(float).mean(axis=0)
df["peak-rpm"]=df["peak-rpm"].fillna(mean_peakrpm)
df.head(5)
```

```python
mode=df["num-of-doors"].value_counts().idxmax()
df["num-of-doors"]=df["num-of-doors"].fillna(mode)
df.head(8)
```

```python
for column in x.columns.values.tolist():
    print(x[column].value_counts())
    print("")
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
#standardization
df["city-mpg"]=235/df["city-mpg"]
df.rename(columns={"city-mpg":"city-L/100km"},inplace=True)
df["highway-mpg"]=235/df["highway-mpg"]
df.rename(columns={"highway-mpg":"highway-L/100km"},inplace=True)
df.head(40)
```

```python
df["length"]=df["length"]/df["length"].max()
df["width"]=df["width"]/df["width"].max()
df["height"]=df["height"]/df["height"].max()
print(df["length"])
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins=np.linspace(min(df["horsepower"]),max(df["horsepower"]),4)
group_names=["low","medium","high"]
df["horsepower-binned"]=pd.cut(df["horsepower"],bins,labels=group_names,include_lowest=True)
df[["horsepower","horsepower-binned"]].head(20)
```

```python

pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python

# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df.head()
```

```python
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
df = pd.concat([df, dummy_variable_2], axis=1)
dummy_variable_2.head()
df.drop('aspiration', axis = 1, inplace=True)
```
